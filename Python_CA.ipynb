##WAP to display the employee’s payslip, Ask the user to enter the following data:
* Employee Name: (sample input – Mark Bate)
* Employee Number: (sample input – 123456789A)
* Week ending: (sample input - 26/01/2018)
* Number of hours worked: (sample input – 42.5)
* Hourly Rate: (sample input – 10.50)
* Overtime Rate: (time-and-a-half as 1.5) --enter only if Number of hours worked>37.5
* Standard Tax Rate: (sample input – 20)
* Overtime Tax Rate: (sample input – 50)--enter only if Number of hours worked>37.5

```
#Author:Rajesh Kumar
#Description: Display Salary Slip
#ChangeLog:
#11-28-2020 RK: Initial Version 
#12-01-2020 RK: Added function
#12-02-2020 RK: Added validation and comment

import datetime


#main function to enter the input with validation 
def payslip():
  while True:
    ename=input("Employee Name : ")
    try:
      assert len(ename)!=0                                                      #Employee name user input validation if left empty
    except:
      print("Name should not be empty.")
      continue
    break
  while True:
    enum=input("Employee Number : ")
    try:
      assert len(enum)!=0                                                       #Employee number user input validation if left empty
    except:
      print("Employee Number should not be empty.")
      continue
    break
  while True:
    wend=input('Week Ending(mm/dd/yyyy): ')
    try:
      wend=datetime.datetime.strptime(wend,'%m/%d/%Y').date()                   #date input validation mm/dd/yyyy
    except:
      print("Invalid Input. Enter again.")
      continue
    break
  while True:
    whrs=input("Number of hours worked: ")
    try:
      whrs=float(whrs)                                                          #working hr validation
      assert whrs >=0 and whrs<=750                                             #working hr should be between 0 to 750
    except:
      print("Enter Worked hrs between 0 to 750. ")
      continue
    break
  while True:
    hrate=input("Hourly Rate: ")
    try:
      hrate=float(hrate)                                                        #hourly rate validation
      assert hrate >=0                                                          #rate should be greater than or equal to 0
    except:
      print("Invalid Input. Enter again")
      continue
    break
  while True:
    if whrs>37.50:
      orate=input("Overtime Rate: ")                                            #user input will be activate only if working hr >37.5
    else:
      break
    try:
      orate=float(orate)                                                        #overtime rate validation
      assert (orate>=0 and orate<=2)                                            #should be between 0 to 2
    except:
      print("Invalid Input. Overtime rate should be less than 2.")
      continue
    break
  while True: 
    tax=input("Standard Tax Rate: ")
    try:
      tax=int(tax)                                                              #standared tax rate validation
      assert tax>=0                                                             #should be greater or equal to 0
    except:
      print("Invalid Input. Enter again")
      continue
    break
  while True:
    if whrs>37.50:
      o_r_tax=input("Overtime Tax Rate: ")                                      #user input will be activate only if working hr >37.5
    else:
      break
    try:
      o_r_tax=int(o_r_tax)                                                      #overtime tax rate validation
      assert o_r_tax>=0                                                         #should be greater or equal to 0
    except:
      print("Invalid Input. Enter again")
      continue
    break
  
  #income function to calculate total income
  def income():
    if whrs<=37.50:                                                             #for <=37.5
      tpay=round(whrs*hrate,2)                                                  #total pay (morking hr * hourly rate) and round it with 2 digit
      return tpay                                                               #return total pay for further use
    else:                                                                       #for <=37.5
      ohrs=round(whrs-37.50,2)                                                  #overtime hr calculation 
      tpay=round(37.50*hrate,2)                                                 #normal pay for 37.5 hr 
      opay=round(ohrs*hrate*orate,2)                                            #overtime pay calculation(overtime hr*normal Rate*times of overtime rate)
      total=round(tpay+opay,2)                                                  #total income (nomal earning + overtime earning)
      return ohrs,tpay,opay,total                                               #return all values for further use
  
  #deduction function to calculate tax deduction
  def deduction():
    if whrs<=37.50:                                                             #tax calculation if working hr<=37.5
      tpay=income()
      stax=round(tpay*tax/100,2)
      return stax
    else:                                                                       #tax calculation if working hr >37.5
      ohrs,tpay,opay,total=income()
      stax=round(tpay*tax/100,2)
      otax=round(opay*o_r_tax/100,2)
      tdct=round(stax+otax,2)
      return stax,otax,tdct

  #function to print/display the salary slip
  def display():
    print("\n{:30} PAYSLIP".format(''))
    print("WEEK ENDING {}".format(datetime.datetime.strftime(wend,'%m/%d/%Y')))
    print("Employeee: {}".format(ename.title()))                                #display name with capitalize first char of each word.
    print("Employee Number: {}".format(enum.upper()))                           #display Employee number with capitalize each word.
    print("\n{:20} Earnings {:20} Deductions".format(' ',' '))
    print("{:20} Hours {:3} Rate {:3} Total".format(' ',' ',' '))
    
    #print the below if working hr <=37.5
    if whrs<=37.50:
      tpay=income()                                                             #calling value from income if working hr <=37.5
      stax=deduction()                                                          #calling value from deduction if working hr <=37.5
      print("Hours (Normal) {0:11,.2f} {1:9,.2f} {2:9,.2f}  Tax @{3:3}% {4:3,.2f} ".format(whrs,hrate,tpay,tax,stax))
      print("\n{0:20} Total Pay: {1:31,.2f}".format(' ',tpay))
      print("{0:20} Total deductions: {1:24,.2f}".format(' ',stax))
      print("{0:20} Net Pay: {1:33,.2f}".format(' ',tpay-stax))
    #print the below if working hr > 37.5
    else:
      ohrs,tpay,opay,total=income()                                             #calling value from income if working hr > 37.5
      stax,otax,tdct=deduction()                                                #calling value from deduction if working hr > 37.5
      print("Hours (Normal) {0:11,.2f} {1:9,.2f} {2:9,.2f}  Tax @{3:3}% {4:3,.2f} ".format(37.50,hrate,tpay,tax,stax))
      print("Hours (Overtime) {0:9,.2f} {1:9,.2f} {2:9,.2f}  Tax @{3:3}% {4:3,.2f} ".format(ohrs,hrate*orate,opay,o_r_tax,otax))
      print("\n{0:20} Total Pay: {1:31,.2f}".format(' ',total))
      print("{0:20} Total deductions: {1:24,.2f}".format(' ',tdct))
      print("{0:20} Net Pay: {1:33,.2f}".format(' ',total-tdct))
      print("\n")
                                                                 
  display()
  

payslip()
```

##WAP which prompts the user for their username in the format Domain Name\Username 
###On entering their domain and username and pressing carriage return, write out to the console window each individual data item

```
#Author: Rajesh Kumar
#description: Program to split domain and username out to the console window each individual data item

#ChangeLog:
#11-25-2020 RK: Initial version
#12-01-2020 RK: added function and comment

#function to split 
def console(txt):                                                               #getting input from function userconsole as txt
  s_char="\\" in txt 
  if s_char==True and txt.count('\\')==1:                                       #validate if the string has "\" and contains only one "\"
    x = txt.split("\\")                                                         #split the string basedon '\'
    #print both the string based on index 0 and 1
    print("\nDomain : {}".format(x[0]))
    print("Username : {}".format(x[1]))
  else:
    print("\nPlease try entering again(domain\\username).\n") 
    userconsole()                                                               #if validation failed re-enter to main function for try-again.
print("#############################\n Welcome to DBS Console \n#############################")

#function to user input
def userconsole():                                                              #function to enter the user input.
  txt=input("Please enter your Username:\n")
  console(txt)                                                                  #calling console function with input        
userconsole()
```

##WAP that prompts the user to enter a number of integer values. 
###The program stores the integers, counts the frequency of each integer and displays the frequency

```
#Author: Rajesh Kumar
#description: Program to counts the frequency of each integer and displays the frequency

#ChangeLog:
#11-29-2020 RK: Initial version
#12-03-2020 RK: added function
#12-03-2020 RK: added validation and comment

#function for user entry
def user_entry():
  print("#############################\n Welcome to DBS Console \n#############################")
  #validate the input based on the given format in try statement
  while True:
    element = input("Input the number of elements to be stored in the list :")  #user input to store the range of number
    try:
      element=int(element)                                                      #user input validation should be integer
      assert element>0                                                          #should be greater than 0
    except:
      print("please enter number/interger value.\n")
      continue
    break
  print("Input {} elements in the list :".format(element))
  num_list = []                                                                 #empty list 
  #loop to enter the value to range entered by user
  for i in range(element):
    while True:
      num = input("element - {} : ".format(i))                                  #enter the number for freequency count
      try:
        num=int(num)                                                            #user input validation should be integer
      except:
        print("Enter number only")
        continue
      break
    num_list.append(num)                                                        #append the value in the list
  return num_list                                                               #return the list for further use


#function for freequency count
def frequency(): 
      num_list=user_entry()                                                     #assigning num_list by calling user_entry function return value
      freq = {}                                                                 #empty dictionary to store freequency
      for num in num_list: 
          if (num in freq): 
              freq[num] += 1
          else: 
              freq[num] = 1
      print("\nThe freequency of all elements of the list:")
      for key, value in freq.items():
        print("{} occurs {} times".format(key, value))                          #print the freequency for each unique number.

if __name__ == "__main__":  
  frequency()                                                                   #calling freequency function.
```
