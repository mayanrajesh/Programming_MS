```

def add_entry():
  print("Add Entry System.\n")
  name=input("Enter Name: ")
  while True:
    account=input("Enter Number: ")
    try:
      account=int(account)                                                          
      assert account >=100                                                        
    except:
      print("Enter valid account number only.")
      continue
    break
  if name.title() in banking:                                                      
    print("{}, already exist.".format(name.title()))                    
  else:
    banking[name.title()] = account                                                  
    print("{}, added in System.".format(name.title()))


def remove():
  print("Delete entry System.\n")
  name=input("Enter Name: ")
  if name.title() in banking:                                                      
    print("\nName: {}".format(name.title()))
    print("Account Number: {}".format(banking[name.title()]))
    while True:
      option=input("\nconfirm if you want to delete {},(yes/no): ".format(name.title()))
      try:
        assert (option.title()=='Yes' or option.title()=='No')                  
      except:
        print("please enter valid Option(yes/no).")
        continue
      break
    if option.title()=='Yes':                                                   
      del banking[name.title()]                                                    
      print("{}, deleted successfully from system.".format(name.title()))
  else:
    print("{}, not found in the syatem.".format(name.title())) 


def deposit():
  print("Add Deposit.\n")
  name=input("Enter Name: ")
  while True:
    amount=input("amount: ")
    try:
      amount=int(amount)                                                          
      assert amount >0                                                        
    except:
      print("Enter valid amount only.")
      continue
    break
  if name.title() in banking:                                                      
    
    payment.setdefault(name.title(),[]).append(amount)  
    print('amount {} added'.format(amount))
    for k, v in payment.items():
      if name.title()==k: 
        print("{}: {}".format(k, sum(v)))
                   
  else:                                                 
    print(" customer not available in System.".format(name.title()))

def withdraw():
  print("withdraw Amount.\n")
  name=input("Enter Name: ")
  while True:
    amount=input("amount: ")
    try:
      amount=int(amount)                                                          
      assert amount >0                                                        
    except:
      print("Enter valid amount only.")
      continue
    break
  if name.title() in banking:
    amount=-amount                                                      
    payment.setdefault(name.title(),[]).append(amount)  
    print('amount {} debited'.format(amount))
    for k, v in payment.items():
      if name.title()==k: 
        print("{}: {}".format(k, sum(v)))                
  else:                                                 
    print(" customer not available in System.".format(name.title()))


def balance_check():
  print("Balance Check.\n")
  name=input("Enter Name: ")
  if name.title() in banking:                                                      
    for k, v in payment.items():
      if name.title()==k: 
        print("{}: {}".format(k, sum(v)))                 
  else:                                                 
    print(" customer not available in System.".format(name.title()))

def banking_sys():                                                  
    print("1. Add customer")
    print("2. Remove customer")
    print("3. Add Deposit")
    print("4. Withdraw Amount")
    print("5. Balance Check")
    print("6. Quit")
    
print("#############################\n Banking system\n#############################")
banking = {}
payment={}
choice = 0                                                                      
banking_sys()
while choice != 6:
  while True:
    choice = input("\nSelect the above option(1-6) : ")
    try:
      choice=int(choice)
      assert (choice in range(1,7))
    except:
      print("enter valid option")
      continue
    break
  if choice==1:
    add_entry()
  elif choice==2:
    remove()
  elif choice==3:
    deposit()
  elif choice==4:
    withdraw()
  elif choice==5:
    balance_check()
  elif choice==6:
    print("Thank you for using bank System.")
  elif choice!=6:
    banking_sys()
```
